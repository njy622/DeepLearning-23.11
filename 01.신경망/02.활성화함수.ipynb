## 활성화(Activation) 함수

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xs = np.linspace(-5, 5, 501)
```

- hraciside step function

```python
def step(x):
    y = np.zeros_like(x)
    y[x > 0] = 1            # x > 0 이면 1, x <= 0 이면 0
    return y
```

```python
plt.figure(figsize=(5,3))
plt.plot(xs, step(xs))
plt.grid(), plt.title('heaviside step function')
plt.show()
```

- sigmoid

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
plt.figure(figsize=(5,3))
plt.plot(xs, sigmoid(xs))
plt.grid(), plt.title('sigmoid function')
plt.show()
```

- hyperbolic tangent

```python
plt.figure(figsize=(5,3))
plt.plot(xs, np.tanh(xs))           # hyperbolic tangent는 함수없이 tanh를 쓰면 됨
plt.grid(), plt.title('hyperbolic tangent function')
plt.show()
```

- relu(rectified linear unit)

```python
def relu(x):
    return np.maximum(0, x)         # relu 이용시: maximum 
```

```python
plt.figure(figsize=(5,3))
plt.plot(xs, relu(xs))           
plt.grid(), plt.title('relu function')
plt.show()
```

- leaky relu (위의 함수보다 살짝 빠진 모양의 그래프)

```python
def leaky_relu(x):
    return np.maximum(0.03*x, x)
```

```python
plt.figure(figsize=(5,3))
plt.plot(xs, leaky_relu(xs))           
plt.grid(), plt.title('leaky_relu function')
plt.show()
```

- softplus

```python
def softplus(x):
    return np.log(1+ np.exp(x))
```

```python
plt.figure(figsize=(5,3))
plt.plot(xs, softplus(xs))      
plt.plot(xs, relu(xs))      
plt.grid(), plt.title('leaky_relu function')
plt.show()
```

- softmax

```python
def softmax(x):
    return np.exp(x) / np.sum(np.exp(x))
```

```python
yhat = np.array([3.0, 1.5, 0.3])
softmax(yhat)
```
