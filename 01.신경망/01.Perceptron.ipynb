```python
# 가중합 구하기
weight = np.array([1,1])
bias = -1
for x1, x2 in [(0,0),(0,1),(1,0),(1,1)]:
    sum = weight[0] * x1 + weight[1] * x2 + bias
    print(f'({x1}, {x2}) :\t {sum}')
```

```python
# Step 활성화 함수
for x1, x2 in [(0,0),(0,1),(1,0),(1,1)]:
    sum = weight[0] * x1 + weight[1] * x2 + bias
    result = 1 if sum > 0 else 0
    print(f'({x1}, {x2}) :\t {sum}\t{result}')
```

```python
# 위의 코드를 좀 더 간결히 나타낼 수 있음
for x in [(0,0),(0,1),(1,0),(1,1)]:
    sum = np.dot(weight, x) + bias
    result = 1 if sum > 0 else 0
    print(f'{x} :\t {sum}\t{result}')
```

```python
def AND_gate(x):                # x는 튜플
    weight = np.array([1,1])
    bias = -1
    sum = np.dot(weight, x) + bias
    return 1 if sum > 0 else 0
```

```python
for x in [(0,0),(0,1),(1,0),(1,1)]:
    print(f'{x} :\t{AND_gate(x)}')
```

- OR gate


```python
def OR_gate(x):
    weight = np.array([2,2])
    bias = -1
    sum = np.dot(weight, x) + bias
    return 1 if sum > 0 else 0
```

```python
for x in [(0,0),(0,1),(1,0),(1,1)]:
    print(f'{x} :\t{OR_gate(x)}')
```

- NAND gete


```python
def NAND_gate(x):
    weight = np.array([-2,-2])
    bias = 3
    sum = np.dot(weight, x) + bias
    return 1 if sum > 0 else 0
```

```python
for x in [(0,0),(0,1),(1,0),(1,1)]:
    print(f'{x} :\t{NAND_gate(x)}')
```

```python
print('X\tOR\tAND\tNAND\t')
for x in [(0,0),(0,1),(1,0),(1,1)]:
    print(f'{x} :\t{OR_gate(x)}\t{AND_gate(x)}\t{NAND_gate(x)}')
```
